# Función de variable aleatoria

Sea X una variable aleatoria normal estándar.

¿Cómo será la nueva variable aleatoria $Y = Y(X) = X^2$?

Es decir, ¿cómo será su función de distribución?

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Simulo la variable aleatoria $X$

```python
x = np.random.randn( 100000 )
```

```python
plt.hist( x, 40, density = 1 );
plt.xlabel( 'x' );
plt.ylabel( 'p(x)' );
```

## Calculo los valores de la nueva variable Y

```python
y = x**2
```

### La teoría dice
$$ p(Y) = p(X(y)) \vert \frac{dX(y)}{dy} \vert = \cdots =\frac{1}{\sqrt{2\pi y}}  e^{-y/2}$$

```python
# La teoría
yt = np.linspace( 0.1, np.max(y), 100 )
xt = np.sqrt( yt )                         # la inversa de y = x^2
pyt = np.exp( - xt**2 / 2 ) / ( np.sqrt( 2 * np.pi ) * xt )
```

## La comparación teoría-experimento

```python
plt.hist( y, 40, density = 1 );
plt.xlabel( 'y = x²' );
plt.ylabel( 'p(y = x²)' );

plt.plot( yt, pyt );
```

## Caso más general: $X \sim N(\mu, \sigma)$ 

```python
mu = 2
sigma = 1
x = mu + sigma * np.random.randn( 100000 )

plt.hist( x, 40, density = 1 );
plt.xlabel( 'x' );
plt.ylabel( 'p(x)' );
```

```python
# los nuevos valores
y = x**2

plt.hist( y, 40, density = 1 );
plt.xlabel( 'y' );
plt.ylabel( 'p(y)' );
```

## Otro ejemplo: $\theta \sim Uniforme(-\pi,\pi), Y = Y(\theta) = cos(\theta)$

```python
x = np.random.rand( 100000 )
theta = ( 2 * x - 1 ) * np.pi 

plt.hist( theta, 40, density = 1 );
plt.xlabel( '$\\theta$' );
plt.ylabel( '$p(\\theta)$' );
```

De acuerdo con la teoría: $$f_Y(y)=f_X(x)\vert \dfrac{dX}{dY} \vert=\frac{1}{\pi}\frac{1}{\sqrt{1-y^2}}$$

```python
y = np.cos( theta )

plt.hist( y, 40, density = 1 );
plt.xlabel( '$\cos(\\theta)$' );
plt.ylabel( '$p(\cos(\\theta))$' );

# la teoría
nv = np.linspace( -.99, .99, 100 )
pnv = 1 / ( np.pi * np.sqrt( 1 - nv**2 ) )

plt.plot( nv, pnv );
```

### Caso práctico: $G = F \frac{M m}{d^2}$

```python
er_F = 0.05; er_M = 0.03; er_m = 0.01; er_d = .05
mu_F = 1e-8;sigma_F = er_F * mu_F
mu_M = 3;sigma_M = er_M * mu_M
mu_m = 0.05;sigma_m = er_m * mu_m
mu_d = 0.032;sigma_d = er_d * mu_d

N = 1000
X = np.random.randn( N, 4 )
F = mu_F + sigma_F * X[:,0]
M = mu_M + sigma_M * X[:,1]
m = mu_m + sigma_m * X[:,2]
d = mu_d + sigma_d * X[:,3]

G = d**2 * F / ( M * m )

plt.hist( G, 40, density = 1 );
plt.xlabel('G');
plt.ylabel('p( G | D )');
```
