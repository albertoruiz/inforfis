# Gráfica paso a paso

## Función matemática

Dado $p$, la probabilidad de un suceso, la función

$$P(n) = 1 - (1-p)^n$$

es la probabilidad de que el suceso ocurra al menos una vez en $n$ repeticiones independientes del experimento.

## Función informática

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def suerte(p,n):
    return 1-(1-p)**n
```

```python
suerte(0.1,5)
```

```python
n = np.arange(10)
```

```python
suerte(0.1,n)
```

## Gráfica

```python
p = 0.15
n = np.arange(1,15+1)
f = suerte(p,n)
```

### Solo el eje y

Si no ponemos las coordenadas x la gráfica usa el índice de cada dato (0,1,2...).

```python
plt.plot(f,'.-');
```

### Indicando el valor real del eje x

En este ejemplo las coordenadas x son también enteros consecutivos, pero empiezan en 1.

```python
plt.plot(n, f,'.-',lw=1);
```

### Formato gráfico

El formato por defecto (png) dibuja los marcadores descentrados. Para mayor calidad elegimos gráficos [svg](https://en.wikipedia.org/wiki/Scalable_Vector_Graphics):

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
plt.plot(n,f,'.-');
```

### Ajustamos los intervalos de representación

```python
plt.plot(n, f,'.-',lw=1);
plt.ylim(0,1);
plt.xlim(0-0.5,max(n)+0.5);
```

### Ponemos los "ticks" que nos gustan en los ejes

```python
plt.plot(n, f,'.-',lw=1);
plt.ylim(0,1);
plt.xlim(0-0.5,max(n)+0.5);
plt.xticks(n);
plt.yticks(np.linspace(0,1,11));
```

### Cambiamos el tamaño y las proporciones de la figura

```python
plt.figure(figsize=(6,6))
plt.plot(n, f,'.-',lw=1);
plt.ylim(0,1);
plt.xlim(0-0.5,max(n)+0.5);
plt.xticks(n);
plt.yticks(np.linspace(0,1,11));
```

### Etiquetamos los ejes y añadimos un grid

```python
plt.figure(figsize=(6,6))
plt.plot(n, f);
plt.plot(n, f,'.',markersize=15,color='red');
plt.ylim(-0.05,1.05);
plt.xlim(0-0.5,max(n)+0.5);
plt.xticks(n);
plt.yticks(np.linspace(0,1,11));
plt.grid(ls='dotted')
#plt.tight_layout()
plt.title('Tentar a la suerte');
plt.xlabel('n')
plt.ylabel('P(n)');
```

### Añadimos leyenda, texto y guardamos en formato pdf

```python
plt.figure(figsize=(6,6))
plt.plot(n, f,'.-',lw=1, markersize=10, label='prob');
plt.ylim(0,1);
plt.xlim(0-0.5,max(n)+0.5);
plt.xticks(n);
plt.yticks(np.linspace(0,1,11));
plt.grid(ls='dotted')
plt.title('Tentar a la suerte');
plt.xlabel('n')
plt.ylabel('P(n)');
plt.text(1,0.8,f'p={p}',fontsize=15);
plt.legend()
plt.savefig('suerte.pdf')
```
