1) Dibuja en una figura la gráfica de la función $f(x)=sin(x)$ (en negro, con trazo grueso), junto con los tres polinomios siguientes (en diferentes colores, con trazo fino). Muestra el intervalo $-1 \leq x \leq 4$ y fija el eje vertical entre -2 y 2.

\begin{aligned}\\
a(x) &= x\\
b(x) &= x - \frac{x^3}{6}\\
c(x) &= x - \frac{x^3}{6} + \frac{x^5}{120}
\end{aligned}


```python
# solución directa, calculando los valores directamente en la función plot

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-1,4,200)

plt.plot(x, np.sin(x), color='black', lw=5)
plt.plot(x, x)
plt.plot(x, x-x**3/6)
plt.plot(x, x-x**3/6+x**5/120)
plt.ylim(-2,2);
```

```python
# otra forma de hacerlo definiendo expresamente cada función

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x)

def a(x):
    return x

def b(x):
    return x - x**3/6

def c(x):
    return  x-x**3/6+x**5/120


x = np.linspace(-1,4,200)

plt.plot(x, f(x), color='black', lw=5, label='f')
plt.plot(x, a(x), label='n=1')
plt.plot(x, b(x), label='n=3')
plt.plot(x, c(x), label='n=5')
plt.ylim(-2,2); plt.legend();
plt.title('Aproximaciones de Taylor');
```

```python
# Esos polinomios son en realidad casos particulares de la
# aproximación de Taylor, que estudiaréis en su momento

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x)

def fact(n):
    return np.product(np.arange(1,n+1))

def p(x,n):
    return sum( [(-1)**(k) * x**(2*k+1) / fact(2*k+1)  for k in range(n)] )

x = np.linspace(-1,4,200)

plt.plot(x, f(x), color='black', lw=5, label='f')
plt.plot(x, p(x,1), label='n=1')
plt.plot(x, p(x,2), label='n=3')
plt.plot(x, p(x,3), label='n=5')
plt.ylim(-2,2); plt.legend();
plt.title('Aproximaciones de Taylor');
```

3) Dibuja una función parecida a la siguiente:



```python
# como hemos comentado en clase, esto se puede conseguir
# fácilmente multiplicando una onda sencilla por una
# camapana de gauss

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-6,6,1000)
y = np.sin(8*x) * np.exp(-x**2/4)
plt.plot(x,y)
plt.axis('off');
```

```python
# podemos ver los ingredientes

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-6,6,1000)
y1 = np.sin(8*x)
y2 = np.exp(-x**2/4)
plt.plot(x,y1,color='gray')
plt.plot(x,y2,color='black')
plt.plot(x,y1*y2,lw=3)
plt.axis('off');
```
