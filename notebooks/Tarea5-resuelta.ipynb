**7)** Dibuja en una figura la gráfica de la función $f(x)=sin(x)$ (en negro, con trazo grueso), junto con los tres polinomios [de Taylor](https://en.wikipedia.org/wiki/Taylor_series)  siguientes(en diferentes colores, con trazo fino). Muestra el intervalo $-1 \leq x \leq 4$ y fija el eje vertical entre -2 y 2.

\begin{aligned}\\
a(x) &= x\\
b(x) &= x - \frac{x^3}{6}\\
c(x) &= x - \frac{x^3}{6} + \frac{x^5}{120}
\end{aligned}


```python
# solución directa, calculando los valores directamente en la función plot

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-1,4,200)

plt.plot(x, np.sin(x), color='black', lw=5)
plt.plot(x, x)
plt.plot(x, x-x**3/6)
plt.plot(x, x-x**3/6+x**5/120)
plt.ylim(-2,2);
```

```python
# otra forma de hacerlo definiendo expresamente cada función

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x)

def a(x):
    return x

def b(x):
    return x - x**3/6

def c(x):
    return  x-x**3/6+x**5/120


x = np.linspace(-1,4,200)

plt.plot(x, f(x), color='black', lw=5, label='f')
plt.plot(x, a(x), label='n=1')
plt.plot(x, b(x), label='n=3')
plt.plot(x, c(x), label='n=5')
plt.ylim(-2,2); plt.legend();
plt.title('Aproximaciones de Taylor');
```

```python
# Esos polinomios son en realidad casos particulares de la
# aproximación de Taylor, que estudiaréis en su momento

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return np.sin(x)

def fact(n):
    return np.product(np.arange(1,n+1))

def p(x,n):
    return sum( [(-1)**(k) * x**(2*k+1) / fact(2*k+1)  for k in range(n)] )

x = np.linspace(-1,4,200)

plt.plot(x, f(x), color='black', lw=5, label='f')
plt.plot(x, p(x,1), label='n=1')
plt.plot(x, p(x,2), label='n=3')
plt.plot(x, p(x,3), label='n=5')
plt.plot(x, p(x,5), label='n=7')
plt.ylim(-2,2); plt.legend();
plt.title('Aproximaciones de Taylor');
```

**9)** Dibuja puntos uniformemente distribuidos dentro del círculo unidad.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# círculo
a = np.linspace(0,np.pi*2,100)
cir_x = np.cos(a)
cir_y = np.sin(a)

# puntos uniformemente distribuidos en el cuadrado
x, y = np.random.rand(2,300)*2 -1

plt.figure(figsize=(6,6))

plt.plot(x,y,'.')

plt.plot(cir_x,cir_y,color='black')
plt.axis('equal');
```

```python
# máscara para seleccionar los puntos que están
# dentro del círculo
ok = x**2 + y**2 < 1

plt.figure(figsize=(6,6))

plt.plot(x[ok],y[ok],'.')

plt.plot(cir_x,cir_y,color='black')
plt.axis('equal');
```

```python
# Si lo hacemos en polares con r uniforme
# los puntos se concentran más cerca del centro
# Para que salgan puntos uniformes hay que
# usar otra distribución de r ¿cuál?

r = np.random.rand(500)
a = np.random.rand(500)*np.pi*2

plt.figure(figsize=(6,6))

plt.plot(r * np.cos(a), r*np.sin(a), '.', color='red')

plt.plot(cir_x,cir_y,color='black')
plt.axis('equal');
```

**12)** Representa gráficamente la densidad de probabilidad normal y su distribución acumulada y muestra sus valores en un punto concreto, como se muestra a continuación. Puedes usar `norm` del módulo `scipy.stats`, que proporciona `norm.pdf()` y `norm.cdf()`. El sombreado se consigue con `plt.fill_between(x,y1,y2, color )`.

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

from scipy.stats import norm

x = np.linspace(-4,4,200)

plt.plot( x, norm.pdf(x) )
plt.plot( x, norm.cdf(x) )


a = 1.2 # por ejemplo

ca = norm.cdf(a)
pa = norm.pdf(a)

plt.plot( a, ca, '.', color='red', markerSize=10 )
plt.plot( a, pa, '.', color='blue', markerSize=10 );

plt.text(2,0.4, f'P(x<{a:0.1f})={ca:0.2f}')
plt.title('Distribución normal')

xa = np.linspace(-4,a,100)
plt.fill_between(xa, norm.pdf(xa), 0, color='lightblue');
```
